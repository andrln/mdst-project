<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
#df = df.groupby(['year']).mean()
df.dropna(subset=['avg_math_8_score'], inplace=True)
df = df.sort_values(by = ['state', 'year'])
df = df.sort_values(by = ['year'])
df["enroll"] = df["enroll"].fillna(df["grades_all_g"])
df['support_services_expenditures_per_student'] = df['support_services_expenditure'] / df['enroll']
df['total_expenditure_per_student'] = df['total_expenditure'] / df['enroll']
df['instruction_expenditure_per_student'] = df['instruction_expenditure'] / df['enroll']
df['total_revenue_per_student'] = df['total_revenue'] / df['enroll']
df['total_expenditure_as_percent_revenue'] = 100*df['total_expenditure_per_student'] / df['total_revenue']
df
```

Dropped na rows for response variable

Sorted dataframe by state and year to get data in proper order by time

Added enroll estimates from tutorial to reduce the amount of missing data

Added support service expenditures per student from tutorial to get a better understanding of money spent

Added the same for total expenditure, instruction expenditure, and revenue

```python
from sklearn.model_selection import train_test_split
```

```python
df.dropna(subset=['year'], inplace=True)
len(df)
df.dropna(subset=['avg_reading_8_score'], inplace=True)
len(df)
df.dropna(subset=['avg_math_4_score'], inplace=True)
len(df)
df.dropna(subset=['instruction_expenditure_per_student'], inplace=True)
len(df)
df.dropna(subset=['support_services_expenditures_per_student'], inplace=True)
len(df)
```

```python
#Since this is time series data, I need to use a sequential train/test split. I use 80% of the data for training.
df_train = df[:int(len(df)*.8)]
df_test = df[int(len(df)*.8):]
```

```python
x_train = df_train[['year','avg_reading_8_score','avg_math_4_score','instruction_expenditure_per_student','support_services_expenditures_per_student']]
#x_train = df_train[['avg_reading_8_score','avg_math_4_score','instruction_expenditure_per_student','support_services_expenditures_per_student']]
#x_train = df_train[['avg_reading_8_score','instruction_expenditure_per_student']]
y_train = df_train['avg_math_8_score']

x_test = df_test[['year','avg_reading_8_score','avg_math_4_score','instruction_expenditure_per_student','support_services_expenditures_per_student']]
#x_test = df_test[['avg_reading_8_score','avg_math_4_score','instruction_expenditure_per_student','support_services_expenditures_per_student']]
#x_test = df_test[['avg_reading_8_score','instruction_expenditure_per_student']]
y_test = df_test['avg_math_8_score']
y_train
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
from xgboost import plot_importance, plot_tree
import xgboost as xgb
```

```python
# create your model here
model = xgb.XGBRegressor(n_estimators=1000)
```

```python
model.fit(x_train, y_train,
        eval_set=[(x_train, y_train), (x_test, y_test)],
        early_stopping_rounds=50,
       verbose=False)
plot = plot_importance(model, height=0.9)
```

```python
df_test['avg_math_8_score_prediction'] = model.predict(x_test)
df_all = pd.concat([df_test, df_train], sort=False).sort_index()
#df_all = df_all.sort_values(by = ['year', 'state'])
plot = df_all[['avg_math_8_score','avg_math_8_score_prediction']].plot(figsize=(15, 7))
df_all
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
# mean error
np.mean(model.predict(x_test)-y_test)
```

```python
# mean absolute error
np.mean(np.abs(model.predict(x_test)-y_test))
```

```python
# root mean squared error -- penalizes large errors
np.mean((model.predict(x_test)-y_test)**2)**0.5
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'instruction_expenditure_per_student'

f = plt.figure(figsize=(12,6))
plt.scatter(x_train[col_name], y_train, color = "red")
plt.scatter(x_train[col_name], model.predict(x_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('avg_math_8_score')
plt.title("Model Behavior On Training Set")
```

```python
col_name = 'instruction_expenditure_per_student'

f = plt.figure(figsize=(12,6))
plt.scatter(x_test[col_name], y_test, color = "blue")
plt.scatter(x_test[col_name], model.predict(x_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('avg_math_8_score')
plt.title("Model Behavior on Testing Set")
```
